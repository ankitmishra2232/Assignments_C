```python
# extract selling date info
data['sale_datetime'] = pd.to_datetime(data['saledate'],utc=True)
```

```python
data['sale_delta']= data['sellingprice']-data['mmr']
```

```python
plt.scatter(data['sale_month'],data['sale_delta'])
plt.ylabel('sale_delta')
plt.xlabel('sale_month ')
plt.show()
```

```python
# filter some outliner

data=data[data['sale_delta']<data['sale_delta'].max()]
data['sale_delta'].mean()
```

```python
x_list =['year','condition','odometer','sale_dayofweek']
y = 'sale_delta'
x_len = len(x_list)
#
fig,ax=plt.subplots(2,2,figsize=(14,8))
for t in range(x_len):
    grid_x = data[x_list[t]]
    grid_y = data[y]
    ax[t//2,t%2].scatter(grid_x,grid_y)
    ax[t//2,t%2].set_title(f'Relation between {grid_x.name} and {grid_y.name}')
plt.show()
```

# Impute Numerical Data

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy = 'mean')
imputer.fit(input_df[numeric_cols])
```

```python
input_df[numeric_cols] = imputer.transform(input_df[numeric_cols])
```

```python
input_df[numeric_cols].describe().loc[['min', 'max']]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(input_df[numeric_cols])
```

```python
input_df[numeric_cols] = scaler.transform(input_df[numeric_cols])
```

```python
input_df[numeric_cols].describe().loc[['min', 'max']]
```

```python
# for categorical_cols

input_df[categorical_cols].nunique().sort_values(ascending=False)
```

```python
from sklearn.preprocessing import OneHotEncoder
# Create the encoder
encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(input_df[categorical_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
len(encoded_cols)
```

```python
input_df[encoded_cols] = encoder.transform(input_df[categorical_cols])
```
